```python
# cropping image to show only parrot

parrotPortrait = img[:,370:950,:]
```

```python
plt.imshow(parrotPortrait)
```

```python
# psuedo color scheme to enhance contrast & visualize data more easily
# selecting channel 0, 1  and 2 one at a time
lum_img1 = parrotPortrait[:, :, 0]
lum_img2 = parrotPortrait[:, :, 1]
lum_img3 = parrotPortrait[:, :, 2]
```

```python
plt.title('When Red channel is selected')
plt.imshow(lum_img1)
plt.colorbar()
```

```python
plt.title('When Green channel is selected')
plt.imshow(lum_img2)
plt.colorbar()
```

```python
plt.title('When Blue channel is selected')
plt.imshow(lum_img3)
plt.colorbar()
```

```python
# setting up a color map
plt.imshow(lum_img1, cmap='hot')
```

```python
# color map can also be set by calling set_cmap() on image object
plt.set_cmap('nipy_spectral')
plt.imshow(lum_img1)
plt.colorbar()
```
