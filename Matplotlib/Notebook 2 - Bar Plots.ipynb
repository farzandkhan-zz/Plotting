```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

from warnings import filterwarnings
filterwarnings('ignore')

import time
```

```python
def dataGenerator():
    ages = np.arange(25, 35, 1)
    salary = np.random.randint(30000,115000,(10))
    pySalary = np.random.randint(28000,150000,(10))
    jsSalary = np.random.randint(25000,90000,(10))
    return ages, np.sort(salary), np.sort(pySalary), np.sort(jsSalary)

agesX, salaryY, pySalaryY, jsSalaryY = dataGenerator()
```

```python
def basicPlot():
    #plt.bar('x', 'height', width=0.8, )
    #there is also an option of log param which when true log=True sets y-axis to be log scale
    # plt.barh() ---- for horizontal bar plots
    plt.bar(agesX, salaryY)
    plt.title('Median salary in (INR) by age')
    plt.xlabel('Ages')
    plt.ylabel('Median salary (INR)')
    plt.legend(['All devs'])
    plt.show()
    
basicPlot()
```

```python
def basicPlot2():
    plt.style.use('fivethirtyeight')
    plt.bar(agesX, salaryY)
    plt.title('Median salary in (INR) by age')
    plt.xlabel('Ages')
    plt.ylabel('Median salary (INR)')
    plt.legend(['All devs'])
    plt.show()
    
basicPlot2()
```

***We can also have both Line and Bar plots in one subplot***

```python
def lineAndBarPlot():
    plt.style.use('fivethirtyeight')
    plt.plot(agesX, pySalaryY, color='#FF0000')
    plt.bar(agesX, salaryY)
    plt.title('Median salary in (INR) by age')
    plt.xlabel('Ages')
    plt.ylabel('Median salary (INR)')
    plt.legend(['All devs'])
    plt.show()
    
lineAndBarPlot()
```

```python
def multipleBars():
    plt.style.use('ggplot')
    x_indexes = np.arange(len(agesX))
    width = 0.25
    plt.bar(x_indexes - 0.25, salaryY, label='All devs', width=width)
    plt.bar(x_indexes, pySalaryY, label='ML', width=width)
    plt.bar(x_indexes + 0.25, jsSalaryY, label='JavaScript', width=width)
    plt.title('Median salary in (INR) by age')
    plt.xlabel('Ages')
    plt.ylabel('Median salary (INR)')
    plt.xticks(ticks=x_indexes, labels=agesX)
    plt.legend()
    plt.tight_layout()
    plt.savefig('multiBars.png')
    plt.show()
    
    
multipleBars()
```
