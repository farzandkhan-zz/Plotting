```python
from matplotlib import pyplot as plt

import numpy as np
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
def dataGenerator():
    ages = np.arange(25, 35, 1)
    salary = np.random.randint(30000,115000,(10))
    pySalary = np.random.randint(28000,150000,(10))
    jsSalary = np.random.randint(25000,90000,(10))
    return ages, np.sort(salary), np.sort(pySalary), np.sort(jsSalary)

agesX, salaryY, pySalaryY, jsSalaryY = dataGenerator()
```

```python
type(agesX)
```

```python
def basicPlot():
    #plt.plot(x,y:'array like or scalar')
    #plot y vs x as lines
    plt.plot(agesX, salaryY)
    plt.axis([22,36,24000,130000]) #[xmin,xmax,ymin,ymax]
    plt.title('Median salary in (INR) by age')
    plt.xlabel('Ages')
    plt.ylabel('Median salary (INR)')
    #legends can be passed as a list in plt.legend() or inside plt.plot(label='All Devs')
    #plt.legend(loc=(0.07,0.05)) for specifying location of legend
    plt.legend(['All devs'])
    plt.show()
    
basicPlot()
```

```python
def multiPlot():
    plt.plot(agesX, salaryY, label='All devs')
    plt.plot(agesX, pySalaryY, label='ML')
    plt.title('Median salary in (INR) by age')
    plt.xlabel('Ages')
    plt.ylabel('Median salary (INR)')
    plt.legend()
    plt.show()
    
multiPlot()
```

**Format Strings**

A format string consists of a part for color, marker and line::

    fmt = '[color][marker][line]'

Each of them is optional. If not provided, the value from the style
cycle is used. Exception: If ``line`` is given, but no ``marker``,
the data will be a line without markers.

**Colors**

The following color abbreviations are supported:

=============    ===============================
character        color
=============    ===============================
``'b'``          blue
``'g'``          green
``'r'``          red
``'c'``          cyan
``'m'``          magenta
``'y'``          yellow
``'k'``          black
``'w'``          white
=============    ===============================

If the color is the only part of the format string, you can
additionally use any  `matplotlib.colors` spec, e.g. full names
(``'green'``) or hex strings (``'#008000'``).

**Markers**

=============    ===============================
character        description
=============    ===============================
``'.'``          point marker
``','``          pixel marker
``'o'``          circle marker
``'v'``          triangle_down marker
``'^'``          triangle_up marker
``'<'``          triangle_left marker
``'>'``          triangle_right marker
``'1'``          tri_down marker
``'2'``          tri_up marker
``'3'``          tri_left marker
``'4'``          tri_right marker
``'s'``          square marker
``'p'``          pentagon marker
``'*'``          star marker
``'h'``          hexagon1 marker
``'H'``          hexagon2 marker
``'+'``          plus marker
``'x'``          x marker
``'D'``          diamond marker
``'d'``          thin_diamond marker
``'|'``          vline marker
``'_'``          hline marker
=============    ===============================

**Line Styles**

=============    ===============================
character        description
=============    ===============================
``'-'``          solid line style
``'--'``         dashed line style
``'-.'``         dash-dot line style
``':'``          dotted line style
=============    ===============================

Example format strings::

    'b'    # blue markers with default shape
    'ro'   # red circles
    'g-'   # green solid line
    '--'   # dashed line with default color
    'k^:'  # black triangle_up markers connected by a dotted line


```python
def multiPlot2():
    # a format string consists of colors, markers and linestyles
    # each of which is optional 
    #plt.plot(c1-.)
    # c for cayan in color
    # 1 for tri-down-mark marker &
    # -. for dash-dot line
    # alternately it can also be written as 
    # plt.plot(color='c', marker='1', linestyle='-.')
    plt.plot(agesX, salaryY, 'k--' , label='All devs')
    plt.plot(agesX, pySalaryY, 'b' , label='ML')
    plt.title('Median salary in (INR) by age')
    plt.xlabel('Ages')
    plt.ylabel('Median salary (INR)')
    plt.legend()
    plt.show()
    
multiPlot2()
```

```python
x = np.linspace(0, 10, 100)
```

```python
def anotherWayofPlotting():
    fig, ax = plt.subplots()#define the number of subplots on the figure
    ax.plot(x, np.sin(x), label='sine')
    #ax.set_title('Simple Sine curve')
    #ax.legend()
    ax.set(xlim=(0, 10), ylim=(-2, 2), xlabel='x', ylabel='sin(x)', title='A Simple Plot');
    
anotherWayofPlotting()  
```

```python
def plottingTwoGraphsInOneFigure():    
    fig, ax = plt.subplots(2)#define the number of subplots on the figure
    ax[0].plot(x, np.sin(x), '-.r', label='sine x')
    ax[1].plot(x, np.cos(x), '--y', label='cos x')
    #ax.set_title('Two graphs in one')
    #ax.set(xlim=(0, 10), ylim=(-2, 2), xlabel='x', ylabel='sin(x)')
    
plottingTwoGraphsInOneFigure()
```

```python
def fourInOne():
    plt.figure()  # create a plot figure
    
    # create the first of two panels and set current axis
    plt.subplot(2, 2, 1) # (rows, columns, panel number)
    plt.plot(x, np.sin(x))
    plt.title('sin X')

    # create the second panel and set current axis
    plt.subplot(2, 2, 2)
    plt.plot(x, np.cos(x))
    plt.title('cosine X')
    
    #third
    plt.subplot(2, 2, 3)
    plt.plot(x, np.tan(x))
    plt.title('tan X')
    
    #fourth
    plt.subplot(2, 2, 4)
    plt.plot(x, np.tanh(x))
    plt.title('tanh X')

fourInOne()
```
