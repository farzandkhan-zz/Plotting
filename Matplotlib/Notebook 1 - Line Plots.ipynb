```python
from matplotlib import pyplot as plt

import numpy as np
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
def dataGenerator():
    ages = np.arange(25, 35, 1)
    salary = np.random.randint(30000,115000,(10))
    pySalary = np.random.randint(28000,150000,(10))
    jsSalary = np.random.randint(25000,90000,(10))
    return ages, np.sort(salary), np.sort(pySalary), np.sort(jsSalary)

agesX, salaryY, pySalaryY, jsSalaryY = dataGenerator()
```

```python
type(agesX)
```

```python
def basicPlot():
    #plt.plot(x,y:'array like or scalar')
    #plot y vs x as lines
    plt.plot(agesX, salaryY)
    plt.title('Median salary in (INR) by age')
    plt.xlabel('Ages')
    plt.ylabel('Median salary (INR)')
    #legends can be passed as a list in plt.legend() or inside plt.plot(label='All Devs')
    #plt.legend(loc=(0.07,0.05)) for specifying location of legend
    plt.legend(['All devs'])
    plt.show()
    
basicPlot()
```

```python
def multiPlot():
    plt.plot(agesX, salaryY, label='All devs')
    plt.plot(agesX, pySalaryY, label='ML')
    plt.title('Median salary in (INR) by age')
    plt.xlabel('Ages')
    plt.ylabel('Median salary (INR)')
    plt.legend()
    plt.show()
    
multiPlot()
```

**Format Strings**

A format string consists of a part for color, marker and line::

    fmt = '[color][marker][line]'

Each of them is optional. If not provided, the value from the style
cycle is used. Exception: If ``line`` is given, but no ``marker``,
the data will be a line without markers.

**Colors**

The following color abbreviations are supported:

=============    ===============================
character        color
=============    ===============================
``'b'``          blue
``'g'``          green
``'r'``          red
``'c'``          cyan
``'m'``          magenta
``'y'``          yellow
``'k'``          black
``'w'``          white
=============    ===============================

If the color is the only part of the format string, you can
additionally use any  `matplotlib.colors` spec, e.g. full names
(``'green'``) or hex strings (``'#008000'``).

**Markers**

=============    ===============================
character        description
=============    ===============================
``'.'``          point marker
``','``          pixel marker
``'o'``          circle marker
``'v'``          triangle_down marker
``'^'``          triangle_up marker
``'<'``          triangle_left marker
``'>'``          triangle_right marker
``'1'``          tri_down marker
``'2'``          tri_up marker
``'3'``          tri_left marker
``'4'``          tri_right marker
``'s'``          square marker
``'p'``          pentagon marker
``'*'``          star marker
``'h'``          hexagon1 marker
``'H'``          hexagon2 marker
``'+'``          plus marker
``'x'``          x marker
``'D'``          diamond marker
``'d'``          thin_diamond marker
``'|'``          vline marker
``'_'``          hline marker
=============    ===============================

**Line Styles**

=============    ===============================
character        description
=============    ===============================
``'-'``          solid line style
``'--'``         dashed line style
``'-.'``         dash-dot line style
``':'``          dotted line style
=============    ===============================

Example format strings::

    'b'    # blue markers with default shape
    'ro'   # red circles
    'g-'   # green solid line
    '--'   # dashed line with default color
    'k^:'  # black triangle_up markers connected by a dotted line


```python
def multiPlot2():
    # a format string consists of colors, markers and linestyles
    # each of which is optional 
    #plt.plot(c1-.)
    # c for cayan in color
    # 1 for tri-down-mark marker &
    # -. for dash-dot line
    # alternately it can also be written as 
    # plt.plot(color='c', marker='1', linestyle='-.')
    plt.plot(agesX, salaryY, 'k--' , label='All devs')
    plt.plot(agesX, pySalaryY, 'b' , label='ML')
    plt.title('Median salary in (INR) by age')
    plt.xlabel('Ages')
    plt.ylabel('Median salary (INR)')
    plt.legend()
    plt.show()
    
multiPlot2()
```

```python
def multiPlot3():
    #plt.style.use('fivethirtyeight')
    plt.plot(agesX, salaryY, color='#444444', linestyle='--', marker='.', linewidth=3, label='All devs')
    plt.plot(agesX, pySalaryY, color='#5a7d9a',marker='o' , label='ML')
    plt.plot(agesX, jsSalaryY, color='#adad3b', marker='s', label='JavaScript')
    plt.title('Median salary in (INR) by age')
    plt.xlabel('Ages')
    plt.ylabel('Median salary (INR)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    
multiPlot3()
```
