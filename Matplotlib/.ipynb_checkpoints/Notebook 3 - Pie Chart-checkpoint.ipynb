```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# do not use pie chart to represent more than 5 variables

def pieChart():
    slices = [60,40,15,10]
    labels=['Sixty', 'Fourty', 'Fifteen', 'Ten']
    colors = ['Blue', 'Red', 'Yellow', 'Green']
    
    #plt.pie(x=array-like, explode=array-like, label=['list'],colors=['list'],
    #                   wedgeprops={'dict'}, startangle='float',
    #                          autopct='%1.1f%% , shadow='True')
    # autopct for auto-percentage - %1.1f%% is the format string
    #explode=[0,0,0,0,0.15,0.15]
    plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})
    plt.title('My awesome Pie Chart')
    plt.tight_layout()
    plt.show()
    
pieChart()
```

```python
def stackOverFlowData():
    slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
    labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
    explode = [0,0,0,0.15,0.15]
    plt.pie(slices[:5], labels=labels[:5], explode=explode, shadow=True, startangle=90, autopct='%1.1f%%')
    plt.title('Language Populatiry')
    plt.tight_layout()
    plt.show()
    
stackOverFlowData()
```
